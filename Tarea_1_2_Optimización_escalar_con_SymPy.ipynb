1. El perímetro de un triángulo isósceles es de $10 \mathrm{cm}$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

Sabiendo que $2l+b=10$, donde $l$ es el lado del triángulo isósceles y $b$ es su base. De esta ecuación, despejamos $l=5-\frac{b}{2}$  
Considerando estos valores, sabemos que por Pitágoras la altura del triángulos igual a:  $h=\sqrt{l^2-\frac{b}{2}^2}$  
Al sustituír $l$ dentro de esta ecuación y desarrollar obtenemos:  
$h=\sqrt{(5-\frac{b}{2})^2-\frac{b}{2}^2}=\sqrt{25-5b+\frac{b}{2}^2-\frac{b}{2}^2}=\sqrt{25-5b}$  
A continuación, sustituímos $r$ de la ecuación del volumen del cono por $\frac{b}{2}$ (también en la ec. de h) y obtenemos lo siguiente:  
$V(r)=\frac{1}{3}\pi r^2 \sqrt{25-10r}$

```python
def V(r):
  return 1/3*np.pi*r**2*(25-10*r)**(1/2)

r=sym.var('r', positive=True)
V(r)
```

```python
dV=V(r).diff(r)
c=sym.solve(dV,r)
c
```

```python
plt.plot(c[0],V(c[0]),'r*', ms=10)
plt.text(c[0]+0.06,V(c[0])+0.1,f'({c[0]:.0f}, {V(c[0]):.2f})', fontsize=10)
r_num=np.linspace(0,10,100)
plt.plot(r_num,V(r_num), lw=2)
plt.xlabel('Radio')
plt.ylabel('Volumen')
plt.grid()
```

**Resultado:**  
Para que el volumen del cono generado sea el máximo ($9.37\mathrm{cm^3}$), el radio del mismo debe de ser $2\mathrm{cm}$.  
Es decir, su base sería de $4\mathrm{cm}$ y sus lados deben medir $3\mathrm{cm}$ c/u.

2. Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

La condición inicial que tenemos es que $2p+t=6$, donde $p$ son los postes y $t$ es el travesaño. De esto despejamos $t=6-2p$  
Lo que queremos maximizar es el área, entonces tomamos en cuenta su fórmulay ahí sustituímos $t$ despejado de la restricción:  
$A=pt=p(6-2p)$

```python
def A(p):
  return p*(6-2*p)

p=sym.var('p', positive=True)
A(p)
```

```python
dA=A(p).diff(p)
c=sym.solve(dA,p)
c
```

```python
plt.plot(c[0],A(c[0]),'r*', ms=10)
plt.text(c[0]+0.1,A(c[0])+0.5,f'({c[0]}, {A(c[0])})', fontsize=10)
p_num=np.linspace(0,6,100)
plt.plot(p_num,A(p_num), lw=2)
plt.xlabel('Poste')
plt.ylabel('Área')
plt.grid()
```

**Resultado:**  
Para que el área de la portería el máximo ($4.5\mathrm{m^2}$), el poste de la portería debe medir $1.5\mathrm{m}$.  
Por lo tanto, su travesaño debería medir $3\mathrm{m}$.

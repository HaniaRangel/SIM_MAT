Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.

1. $f(x)=x+\frac{1}{x^2}$

Importamos todas las librerías que usaremos/podemos utilizar en los problemas

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=sym.var('x')
def f(x):
    return x+1/x**2
df=f(x).diff(x)
c=sym.solve(df,x)
c
```

Dentro del resultado se obtienen 3 valores, de los cuales 2 son complejos.
Para corregir esto indicamos que la variable debe ser real añadiendo `
real=True `

```python
x=sym.var('x',real=True)
def f(x):
    return x+1/x**2
df=f(x).diff(x)
c=sym.solve(df,x)
#c[0]=2**(1/3)
#float(c[0])=1.2599210498948732

x_num=np.linspace(1,1.5,100)
plt.plot(x_num,f(x_num))
plt.plot(c[0],f(c[0]),'r*')
plt.grid()
```

R: $2^{1/3}$ es un punto crítico minimo global

2. $f(x)=1260+117x-9x^2$

```python
def f2(x):
    return 1260+117*x-9*x**2
df2=f2(x).diff(x)
c2=sym.solve(df2,x)
#c2[0]=13/2

ddf2=df2.diff(x)
#ddf2=-18 <- máx

x2_num=np.linspace(3,10,100)
plt.plot(x2_num,f2(x2_num))
plt.plot(c2[0],f2(c2[0]),'r*')
plt.grid()
```

R: $13/2$ es un punto crítico máximo global

3. $f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
def f3(x):
  return 5+x+4/x
df3=f3(x).diff(x)
c3=sym.solve(df3,x)
#c3=[-2, 2] <- solo tomamos x>0
#c3[1]=2

x3_num=np.linspace(1.5,2.5,100)
plt.plot(x3_num,f3(x3_num))
plt.plot(c3[1],f3(c3[1]),'r*')
plt.grid()
```

R: $2$ es un punto crítico minimo global para $x>0$


4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

```python
def f4(x):
  return 15*x/2*sym.sqrt(x/(x-30))
df4=f4(x).diff(x)
c4=sym.solve(df4,x)
#c4=[0, 45] <- solo c4[1] entra en el dominio de x>0
```

Como se usó sym dentro de la función $f4$ solo acepta argumentos simbólicos, se necesita generar otra función para que acepte el arreglo `array`

```python
f4_num=sym.lambdify(x, f4(x), 'numpy')

x4_num=np.linspace(42, 48, 100)
plt.plot(x4_num, f4_num(x4_num))
plt.plot(float(c4[1]), float(f4(c4[1])), 'r*')
plt.grid()
```

R: $45$ es un punto crítico minimo global para $x>0$

5. $f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
def f5(x):
  return 6-2*x+sym.sqrt(18-2*x**2)
df5=f5(x).diff(x)
c5=sym.solve(df5,x)
#c5[0]=-sqrt(6)

f5_num=sym.lambdify(x, f5(x), 'numpy')

x5_num=np.linspace(-3, 3, 100) #dominio de la función
plt.plot(x5_num, f5_num(x5_num))
plt.plot(float(c5[0]), float(f5(c5[0])), 'r*')
plt.grid()
```

R: $-\sqrt{6}$ es un punto crítico máximo global para $-3\leq x\leq 3$

6. $f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
def f6(x):
  return 1/4*x**4+1/3*x**3-x**2
df6=f6(x).diff(x)
c6=sym.solve(df6,x)
#c6=[-2, 0, 1] <- todos están dentro del intérvalo

x6_num=np.linspace(-3, 1.7, 100)
plt.plot(x6_num, f6(x6_num))
plt.plot(c6[0], f6(c6[0]), 'r*')
plt.plot(c6[1], f6(c6[1]), 'r*')
plt.plot(c6[2], f6(c6[2]), 'r*')
plt.text(c6[0]-0.5,f6(c6[0])+0.2,f'({c6[0]:.2f},{f6(c6[0]):.2f})', fontsize=14)
plt.text(c6[1]-0.5,f6(c6[1])+0.2,f'({c6[1]:.2f},{f6(c6[1]):.2f})', fontsize=14)
plt.text(c6[2]-0.5,f6(c6[2])+0.2,f'({c6[2]:.2f},{f6(c6[2]):.2f})', fontsize=14)
plt.grid()
```

**R:**  
$-2$ es un punto crítico mínimo global  
$0$ es un punto crítico máximo local  
$1$ es un punto crítico mínimo local

7. $f(x)=100(1+\cos(x)\sin(x))$, para $x\in[0.\frac{\pi}{2}]$

```python
def f7(x):
  return 100*(1+sym.cos(x)*sym.sin(x))
df7=f7(x).diff(x)
c7=sym.solve(df7,x)
#c7=[-pi/4, pi/4] <- solo c7[1] está dentro del dominio

f7_num=sym.lambdify(x, f7(x), 'numpy')

x7_num=np.linspace(0, np.pi/2, 100) #dominio de la función
plt.plot(x7_num, f7_num(x7_num))
plt.plot(float(c7[1]), float(f7(c7[1])), 'r*')
plt.grid()
```

R: $\pi/4$ es un punto crítico máximo global para $x\in[0.\frac{\pi}{2}]$

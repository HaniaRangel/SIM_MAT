Importar los precios de Amazon desde Yahoo finance de los últimos 10 años y simular cien mil escenarios de los próximos 30 días para predecir la probabilidad de que el precio actual supere el 8%.

```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
#Descargar marca de interés
#AMZN - Amazon
data=yf.download('AMZN',start='2015-11-04',end='2025-11-04')['Close']#<- indica la columna a extraer
#Extraemos los precios al cierre de los últimos 10 años
data.plot(grid=True)
```

```python
delta=data.pct_change().dropna()
delta.plot(grid=True) # se consideran medias constantes
```

```python
delta.hist(bins=50,grid=True) # resultados centrados
```

```python
#Simular los deltas
N_dias=30
N_escenarios=100000
delta_AMZN=np.random.normal(delta['AMZN'].mean(),delta['AMZN'].std(),(N_dias,N_escenarios))

#Asociamos a los días simulados (creamos un rango de tiempo)
rango_tiempo=pd.date_range('2025-11-04',periods=N_dias)
```

```python
# Tabla N_escenarios x N_dias
delta_AMZN=pd.DataFrame(delta_AMZN,index=rango_tiempo)
delta_AMZN
```

```python
#Proyección de precios (Fórmula de P_k)
precios_AMZN=data['AMZN'].iloc[-1]*(1+delta_AMZN).cumprod()
precios_AMZN.iloc[:2,:5]
```

```python
#Probabilidad de que supere el 8%
B_AMZN=precios_AMZN>data['AMZN'].iloc[-1]*1.08 #variable de bernaulli (cuantos escenarios satisfacen la condición buscada)
probab_AMZN=B_AMZN.mean(axis=1)
probab_AMZN.plot(grid=True)
```

Basandonos en las tendencias de los precios de Amazon en los últimos $10$ años y simulando $100,000$ escenarios con simulación Montecarlo, sabemos que la probabilidad de que en $30$ días el precio actual supere el $8\%$ es de $31.99\%$
